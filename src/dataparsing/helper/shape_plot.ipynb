```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
csv_path="/home/amanandhar/ASCC/abacus_file/mesh.csv"
df=pd.read_csv(csv_path)
df.head()
```

```python
csv_path="/home/amanandhar/ASCC/abacus_file/temp2.csv"
df_temp=pd.read_csv(csv_path)
df_temp.tail()
```

```python
df_temp_avg=df_temp
df_temp_avg=df_temp_avg.groupby('Node')['Temp'].agg(avg_temp='mean').reset_index()
```

```python
merged_df=pd.merge(df,df_temp_avg,on='Node',how='inner')
merged_df.head()
```

```python
# Create a 3D scatter plot with intensity based on the 'temp' column
fig = px.scatter_3d(merged_df, x='x', y='y', z='z', color='avg_temp', title='3-D shape with avg_temp')
fig.update_traces(marker=dict(size=8))  # Customize marker size

# Show the plot (optional)
fig.show()
```
